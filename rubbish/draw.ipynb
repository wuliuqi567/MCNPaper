```python

import os
import matplotlib.pyplot as plt
import numpy as np


file = './testsatevideo.log'
bps = []
with open(file, 'r') as f :
    lines = f.readlines()
    id = 0
    for line in lines:
        line = line.split()[1].split('/')[4]
        # print(type(line))
        # print(line)
        bps_str = line[6:-3]
        bps.append(int(bps_str))


    print(bps)

x = range(len(bps)//2)

plt.figure()
plt.plot(x, bps[:len(bps)//2])
font2 = {'family':'Time New Roman','weight' : 'normal','size'   : 18,}

# my_x_ticks = np.arange(-5, 5, 0.5)
# my_y_ticks = np.arange(0, 3936261, 10000)
# # plt.xticks(my_x_ticks)
# plt.yticks(my_y_ticks)

plt.xlabel('num', font2)
plt.ylabel('bps',font2)
plt.figure()
plt.plot(x, bps[len(bps)//2:])
plt.xlabel('num',font2)
plt.ylabel('bps',font2)
plt.draw()
```
