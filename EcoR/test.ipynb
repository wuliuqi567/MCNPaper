```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

nodes = [0,1,2,3,4]


print(max(nodes))
edges = [(0,1),(0,2),(0,3), (1,2),(1,3),(2,3),(0,4), (2,4)]
G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
# print(list(nx.dfs_edges(G,0))) #1)
a = list(nx.all_simple_paths(G,0,4))#1
print(a)
b = list(nx.all_simple_paths(G,0,4,cutoff=3))#2
print(b)
c = list(nx.all_simple_paths(G,0,4,cutoff=2))#3
print(c)
#绘制无向图
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
```
